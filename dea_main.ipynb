```python
# It shows that the distribution is not normal, let's check the outliers, how many they are
# Boxplot in order to see if there are outliers
plt.boxplot(internet_penetration_perstate['Accesos por cada 100 hogares'])
plt.show()
```

```python
# Due to there are outliers, let's drop them
# Preparation of quartils
Q1 = internet_penetration_perstate["Accesos por cada 100 hogares"].quantile(0.25)
Q3 = internet_penetration_perstate["Accesos por cada 100 hogares"].quantile(0.75)
IQR = Q3 - Q1

# Filter outliers
internet_penetration_perstate[(internet_penetration_perstate["Accesos por cada 100 hogares"] >= Q1 - 1.5 * IQR) & (internet_penetration_perstate["Accesos por cada 100 hogares"] <= Q3 + 1.5 * IQR)]
```

```python
# Lets prepare the information for a bar plot, grouping the dataset by "Provincia"
internet_penetration__perstate_filtered= internet_penetration_perstate.groupby('Provincia')['Accesos por cada 100 hogares'].mean().reset_index().sort_values('Accesos por cada 100 hogares')
```

```python
#Check which states have the lowest penetration
fig, ax = plt.subplots()
x= internet_penetratiointernet_penetration__perstate_filteredn_filtered['Provincia']
y= internet_penetratiointernet_penetration__perstate_filteredn_filtered['Accesos por cada 100 hogares']
ax.bar(x,y)
ax.set_title('Accesos a internet por provincia')
ax.set_xlabel('Provincia')
ax.set_ylabel('Accesos por cada 100 hogares')
ax.set_xticklabels(internet_penetratiointernet_penetration__perstate_filteredn_filtered['Provincia'],rotation= 90)
plt.show()
```

` Let's check this dataset "ConectividadalserviciodeInternet_2776151681933683561"`: This dataset shows a list of states and cities that has internet connection per service

```python
# Import dataset
connection_perservice= pd.read_csv(r'Datasets_import\ConectividadalserviciodeInternet_2776151681933683561.csv')
# Quick view
print(connection_perservice.shape)
connection_perservice.head()
```

```python
# Let's check the types
connection_perservice.info()
```

```python
connection_perservice.describe()
```

```python
# Lets group by "Provincia", sum the "Poblacion" column and count the "SI" for the rest of the columns.
# We drop "link", "latitud" and "Longitud" columns due to they do not apport for the analysis
connection_perservice = connection_perservice.groupby('Provincia').agg({'Poblacion': 'sum', 
                                                                         'ADSL': lambda x: sum(x=='SI'), 
                                                                         'CABLEMODEM': lambda x: sum(x=='SI'), 
                                                                         'DIALUP': lambda x: sum(x=='SI'), 
                                                                         'FIBRAOPTICA': lambda x: sum(x=='SI'), 
                                                                         'SATELITAL': lambda x: sum(x=='SI'), 
                                                                         'WIRELESS': lambda x: sum(x=='SI'), 
                                                                         'TELEFONIAFIJA': lambda x: sum(x=='SI'), 
                                                                         '3G': lambda x: sum(x=='SI'), 
                                                                         '4G': lambda x: sum(x=='SI')})

connection_perservice= connection_perservice.reset_index()
connection_perservice
```

```python
# Let's create a bar plot

data = connection_perservice
df = pd.DataFrame(data)

df_plot = df[['Provincia',	'ADSL',	'CABLEMODEM', 'DIALUP', 'FIBRAOPTICA', 'SATELITAL', 'WIRELESS', 'TELEFONIAFIJA', '3G','4G']]
df_plot = df_plot.set_index('Provincia')

ax = df_plot.plot(kind='bar', stacked=True, figsize=(10, 6))

ax.set_title('Internet access by states in Argentina')
ax.set_xlabel('State')
ax.set_ylabel('Access type')
```

```python
# Let's see if there is any correlation between variables
sns.heatmap(connection_perservice.drop(columns=["Provincia","Poblacion"]), cmap="YlGnBu")
plt.show
```

```python
connection_perservice.to_csv('Datasets_export/connection_perservice.csv')
```

* Buenos Aires is the state better connected to internet in every service. This could be a consequence of the population.

```python
#linefig to see the tendance for diferent variables of the "speed_access" dataset
x= speed_access_year['Año']
y1= speed_access_year['Hasta 512 kbps']
y2= speed_access_year['Entre 512 Kbps y 1 Mbps']
y3= speed_access_year['Entre 1 Mbps y 6 Mbps']
y4= speed_access_year['Entre 6 Mbps y 10 Mbps']
y5= speed_access_year['Entre 10 Mbps y 20 Mbps']
y6= speed_access_year['Entre 20 Mbps y 30 Mbps']
y7= speed_access_year['Más de 30 Mbps']

fig, ax1 = plt.subplots(figsize= (8,5))
ax1.set_xlabel('Periodo')
ax1.set_ylabel('Hasta 512 kbps')
ax1.set_xticklabels(speed_access_year['Año'],rotation= 90)
ax1.plot(x, y1, color = "r")
ax1.tick_params(axis='y', labelcolor='r')

ax2= ax1.twinx()
ax2.set_ylabel('Entre 512 Kbps y 1 Mbps')
ax2.plot(x,y2,color= "b")
ax2.tick_params(axis="y",labelcolor= "b")

ax3= ax1.twinx()
ax3.set_ylabel('Entre 1 Mbps y 6 Mbps')
#ax3.spines["right"].set_position(("axes", 1.1))
ax3.plot(x,y3,color= "g")
ax3.tick_params(axis="y",labelcolor= "g")

ax4= ax1.twinx()
ax4.set_ylabel('Entre 6 Mbps y 10 Mbps')
ax4.plot(x,y4,color="m")
ax4.tick_params(axis="y",labelcolor= "m")

ax5= ax1.twinx()
ax5.set_ylabel('Entre 10 Mbps y 20 Mbps')
ax5.plot(x,y5,color="y")
ax5.tick_params(axis="y",labelcolor= "y")

ax6= ax1.twinx()
ax6.set_ylabel('Entre 20 Mbps y 30 Mbps')
ax6.plot(x,y6,color="c")
ax6.tick_params(axis="y",labelcolor= "c")

ax7= ax1.twinx()
ax7.set_ylabel('Más de 30 Mbps')
ax7.plot(x,y7,color="k")
ax7.tick_params(axis="y",labelcolor= "k")

plt.show()
```

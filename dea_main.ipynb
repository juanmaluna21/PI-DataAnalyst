Now let's check "Internet Penetration"

```python
df_internet_penetration= pd.read_csv(r'Datasets\Internet_Penetracion.csv')
print(df_internet_penetration.shape)
df_internet_penetration.head()
```

```python
df_internet_penetration.info()
```

```python
#Transformations to be able to make graphs
df_internet_penetration['Accesos por cada 100 hogares']= df_internet_penetration['Accesos por cada 100 hogares'].str.replace(",",'.')
df_internet_penetration['Accesos por cada 100 hogares']= df_internet_penetration['Accesos por cada 100 hogares'].astype('float64')
```

```python
#Check some metrics
df_internet_penetration.describe()
```

```python
# Boxplot in order to see if there are outliers
plt.boxplot(df_internet_penetration['Accesos por cada 100 hogares'])
plt.show()
```

```python
# Due to there are outliers, let's drop them
# Preparation of quartils
Q1 = df_internet_penetration["Accesos por cada 100 hogares"].quantile(0.25)
Q3 = df_internet_penetration["Accesos por cada 100 hogares"].quantile(0.75)
IQR = Q3 - Q1
# Filter outliers
df_internet_penetration[(df_internet_penetration["Accesos por cada 100 hogares"] >= Q1 - 1.5 * IQR) & (df_internet_penetration["Accesos por cada 100 hogares"] <= Q3 + 1.5 * IQR)]
```

```python
df_internet_penetration_filtered= df_internet_penetration.groupby('Provincia')['Accesos por cada 100 hogares'].mean().reset_index().sort_values('Accesos por cada 100 hogares')
```

```python
df_internet_penetration_filtered= df_internet_penetration_filtered.head(10)
```

```python
#Check which states have the lowest penetration
fig, ax = plt.subplots()
x= df_internet_penetration_filtered['Provincia']
y= df_internet_penetration_filtered['Accesos por cada 100 hogares']
ax.bar(x,y)
ax.set_title('Accesos a internet por provincia')
ax.set_xlabel('Provincia')
ax.set_ylabel('Accesos por cada 100 hogares')
ax.set_xticklabels(df_internet_penetration_filtered['Provincia'],rotation= 90)
plt.show()
```

```python
df_internet_penetration= df_internet_penetration.reset_index()
```

```python
df_internet_penetration.to_csv('Datasets_export/internet_penetration.csv')
```

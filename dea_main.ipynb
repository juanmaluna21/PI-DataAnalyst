`Now let's check "Internet Penetration"`

```python
df_internet_penetration= pd.read_csv(r'Datasets_import\Internet_Penetracion.csv')
print(df_internet_penetration.shape)
df_internet_penetration.head()
```

```python
df_internet_penetration.info()
```

```python
#Transformations to be able to make graphs
df_internet_penetration['Accesos por cada 100 hogares']= df_internet_penetration['Accesos por cada 100 hogares'].str.replace(",",'.')
df_internet_penetration['Accesos por cada 100 hogares']= df_internet_penetration['Accesos por cada 100 hogares'].astype('float64')
```

```python
#Check some metrics
df_internet_penetration.describe()
```

```python
# Let's check its distribution
access_100h= df_internet_penetration['Accesos por cada 100 hogares']
plt.hist(access_100h, bins=20)
plt.xlabel('Acceso por cada 100 hab')
plt.ylabel('Frequence')
plt.title('Argentina internet access distribution')
plt.show()
#It shows that the distribution is not normal, let's check the outliers on next step
```

```python
# Boxplot in order to see if there are outliers
plt.boxplot(df_internet_penetration['Accesos por cada 100 hogares'])
plt.show()
```

```python
# Due to there are outliers, let's drop them
# Preparation of quartils
Q1 = df_internet_penetration["Accesos por cada 100 hogares"].quantile(0.25)
Q3 = df_internet_penetration["Accesos por cada 100 hogares"].quantile(0.75)
IQR = Q3 - Q1
# Filter outliers
df_internet_penetration[(df_internet_penetration["Accesos por cada 100 hogares"] >= Q1 - 1.5 * IQR) & (df_internet_penetration["Accesos por cada 100 hogares"] <= Q3 + 1.5 * IQR)]
```

```python
df_internet_penetration_filtered= df_internet_penetration.groupby('Provincia')['Accesos por cada 100 hogares'].mean().reset_index().sort_values('Accesos por cada 100 hogares')
```

```python
df_internet_penetration_filtered= df_internet_penetration_filtered.head(10)
```

```python
#Check which states have the lowest penetration
fig, ax = plt.subplots()
x= df_internet_penetration_filtered['Provincia']
y= df_internet_penetration_filtered['Accesos por cada 100 hogares']
ax.bar(x,y)
ax.set_title('Accesos a internet por provincia')
ax.set_xlabel('Provincia')
ax.set_ylabel('Accesos por cada 100 hogares')
ax.set_xticklabels(df_internet_penetration_filtered['Provincia'],rotation= 90)
plt.show()
```

```python
df_internet_penetration= df_internet_penetration.reset_index()
```

```python
df_internet_penetration.to_csv('Datasets_export/internet_penetration.csv')
```

From this analysis we can see that we have many states with low internet access. Between them, we can see:
"Mendoza", "Catamarca", "Jujuy","Corrientes","Misiones","Chaco","San Juan", "Santa Cruz","Santiago del Estero" and "Formosa" as the lowest.

```python
# Let's drop unnecessary column 
speed_access = speed_access.drop(columns='Unnamed: 11')
# Let's drop all the "."
speed_access= speed_access.applymap(lambda s: s.replace(".","") if isinstance(s, str) else s)
# Now we have change the type from "object" to "float"
speed_access = speed_access.applymap(lambda s: float(s.replace(".", "")) if isinstance(s, str) and s.replace(".", "").isdigit() else s)
```

```python
speed_access.info()
```

```python
speed_access_year= speed_access.groupby('AÃ±o')[speed_access.columns[2:]].apply(lambda x: x.sum())
speed_access_year= speed_access_year.reset_index()
speed_access_year.head()
```

```python
#linefig to see the tendance for diferent variables of the "speed_access" dataset
x= speed_access_year['AÃ±o']
y1= speed_access_year['Hasta 512 kbps']
y2= speed_access_year['Entre 512 Kbps y 1 Mbps']
y3= speed_access_year['Entre 1 Mbps y 6 Mbps']
y4= speed_access_year['Entre 6 Mbps y 10 Mbps']
y5= speed_access_year['Entre 10 Mbps y 20 Mbps']
y6= speed_access_year['Entre 20 Mbps y 30 Mbps']
y7= speed_access_year['MÃ¡s de 30 Mbps']

fig, ax1 = plt.subplots(figsize= (8,5))
ax1.set_xlabel('Periodo')
ax1.set_ylabel('Hasta 512 kbps')
ax1.set_xticklabels(speed_access_year['AÃ±o'],rotation= 90)
ax1.plot(x, y1, color = "r")
ax1.tick_params(axis='y', labelcolor='r')

ax2= ax1.twinx()
ax2.set_ylabel('Entre 512 Kbps y 1 Mbps')
ax2.plot(x,y2,color= "b")
ax2.tick_params(axis="y",labelcolor= "b")

ax3= ax1.twinx()
ax3.set_ylabel('Entre 1 Mbps y 6 Mbps')
#ax3.spines["right"].set_position(("axes", 1.1))
ax3.plot(x,y3,color= "g")
ax3.tick_params(axis="y",labelcolor= "g")

ax4= ax1.twinx()
ax4.set_ylabel('Entre 6 Mbps y 10 Mbps')
ax4.plot(x,y4,color="m")
ax4.tick_params(axis="y",labelcolor= "m")

ax5= ax1.twinx()
ax5.set_ylabel('Entre 10 Mbps y 20 Mbps')
ax5.plot(x,y5,color="y")
ax5.tick_params(axis="y",labelcolor= "y")

ax6= ax1.twinx()
ax6.set_ylabel('Entre 20 Mbps y 30 Mbps')
ax6.plot(x,y6,color="c")
ax6.tick_params(axis="y",labelcolor= "c")

ax7= ax1.twinx()
ax7.set_ylabel('MÃ¡s de 30 Mbps')
ax7.plot(x,y7,color="k")
ax7.tick_params(axis="y",labelcolor= "k")

plt.show()
```
